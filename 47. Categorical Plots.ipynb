```python
import seaborn as sns
import pandas as pd
%matplotlib inline
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data = tips)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill', data = tips, estimator = np.std)    # estimator에 standard deviation 추가
```

```python
sns.countplot(x = 'sex', data = tips)       # 단순한 count 임. 남자가 여자보다 더 팁 많이줌
```

```python
sns.boxplot(x='day', y='total_bill', data = tips, hue = 'smoker')
```

## x 는 total bill 이고 날짜에 따라 나뉜거임. dots 는 outlier 

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue='sex', split = True)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips)    # 얼마나 스택 되어있는지 모를땐 jitter=True 추가
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = True)
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips)    # swarm 은 violin plot 과 strip plot 합친 것
```

```python
sns.factorplot(x = 'day', y = 'total_bill', data = tips, kind = 'bar')
```
