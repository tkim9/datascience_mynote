# Matplotlib 

### matplotlib.org/gallery.html 여기 가면 차트 같은거 다 그림으로 볼 수 있음

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0 , 5, 11)
```

```python
x
```

```python
y = x ** 2
y
```

```python
# Functional

plt.plot(x,y)
plt.show()              # print()같이 주피터 노트북에선 이거 필요 없지만 다른 것 쓸땐 꼭 써야됨
```

```python
# Labels
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# 하나에 두개의 그래프를 넣고 싶은 경우

plt.subplot(1,2,1)     #   subplot(nrows, ncols, plot_number)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Method - 이게 더 많은 컨트롤 가능

fig = plt.figure()         # 여기서 fig는 그냥 blank canvas 라고 생각 

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])       # 0 에서 1 사이의 [left, bottom, width, height]
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])                   # [왼쪽으로 부터, 밑에서 부터, 세로길이, 가로길이]

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig = plt.figure()
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 2)

plt.tight_layout()         # 그래프들 서로 안겹치게 해결
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
axes              # arrays of matplotlib
```

## 결국 plt.subplots 은 plt.figure 을 통해서 수동으로 설정하는 것을 layout 에 맞게 자동으로 해주는 거임

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()      # 이거는 항상 해주는 것이 좋음

# 결국 list 다루듯이 하면됨
```

# Figure Size and DPI 

```python
fig = plt.figure(figsize=(8,2), dpi = 100)       # figsize는 inches 임. dpi 는 pixels per axes
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_pictue.png', dpi = 200)                  # 그림파일로 저장하기
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Squared')

ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')

ax.plot(x,x**3, label = 'X Cubed')

# you now need legend. 2 step 필요함: label 해주고 legend() 실행

ax.legend(loc = (0.1,0.7))                # location 도 임의로 설정해줄 수 있음. 사이트 가서 더보기. 아무것도 안넣으면 알아서 해줌
```

# Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00', linewidth = 3, alpha = 0.5)              # 그냥 색 이름이나 RGB Hex code 사용 가능 
                                                        # linewidth (또는 lw) 는 default 가 1 이고 크면 더 두꺼워짐
                                                        # alpha 는 transparency of the line
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00', linewidth = 3, linestyle = '--')   # linestyle 는 여러가지 있으니 검색해보기
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00', linewidth = 3, linestyle = '-', marker='o', markersize = 10,
       markerfacecolor = 'yellow', markeredgewidth = 3, markeredgecolor = 'green')      # marks each point

# 나중에 검색해보기
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00', linewidth = 3, linestyle = '--') 

ax.set_xlim([0,1])
ax.set_ylim([0,2])              # [lowerbound, upperbound] 를 통해서 zoom in 가능
```
