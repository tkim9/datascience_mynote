```python
import seaborn as sns
import pandas as pd
%matplotlib inline
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
tips.head()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex')     # separate by sex using hue
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['o', 'v'],
          scatter_kws={'s':100})     # size 를 dictionary 형식으로 한 것임
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'sex')     # sex로 column을 나눔
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'sex', row ='time')     # row 도 가능
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'day', hue = 'sex', 
          aspect = 0.6, size = 8)        

# aspect 는 width / height 임
```
